```
idades = [25, 45, 18, 30, 22, 60, 52]


media = sum(idades)/len(idades)
print(f"A média das idades é de:{media}")
print("As idades maiores do que a média são:")
for idade in idades:
  if idade > media:
    print(idade, end="; ")
```
