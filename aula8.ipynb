```
# Dica: você pode verificar se uma chave existe usando o operador 'in'
# if nome_produto in dicionario:

# Seu código aqui


produtos = {
    "banana": 2.3,
    "maçã": 3.5,
    "ameixa": 5.0
}

nome_produto=input("Digite o produto a ser consutado:").lower()
if nome_produto in produtos:
  print(f"o valor é R$:{produtos[nome_produto]}")
else:
  print("Produto não encontrado")
```
