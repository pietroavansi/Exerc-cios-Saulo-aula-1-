```
def calcular_anos_populacao():
    while True:
        try:
            pop_a = float(input("Digite a população inicial do país A: "))
            if pop_a <= 0:
                print("A população deve ser um número positivo.")
                continue

            taxa_a_percent = float(input("Digite a taxa anual de crescimento do país A (em %): "))
            if taxa_a_percent < 0:
                print("A taxa de crescimento não pode ser negativa.")
                continue
            taxa_a = taxa_a_percent / 100

            pop_b = float(input("Digite a população inicial do país B: "))
            if pop_b <= 0:
                print("A população deve ser um número positivo.")
                continue

            taxa_b_percent = float(input("Digite a taxa anual de crescimento do país B (em %): "))
            if taxa_b_percent < 0:
                print("A taxa de crescimento não pode ser negativa.")
                continue
            taxa_b = taxa_b_percent / 100

            if pop_a >= pop_b and taxa_a <= taxa_b:
                print("A população A já é maior ou igual à B e/ou a taxa de crescimento de A é menor ou igual à de B. Não haverá ultrapassagem.")
                anos = 0 # or handle as appropriate, maybe -1 to indicate no overtake
            elif pop_a < pop_b and taxa_a <= taxa_b:
                print("A população A nunca ultrapassará a população B com essas taxas de crescimento, pois A é menor e cresce mais devagar ou igual.")
                anos = -1 # Indicate no overtake possible
            else:
                anos = 0
                pop_a_temp = pop_a
                pop_b_temp = pop_b
                while pop_a_temp < pop_b_temp:
                    pop_a_temp += pop_a_temp * taxa_a
                    pop_b_temp += pop_b_temp * taxa_b
                    anos += 1

            if anos == -1:
                print(f"Não é possível que a população de A ultrapasse a de B com os dados fornecidos.")
            else:
                print(f"\nSerão necessários {anos} anos para que a população do país A ultrapasse ou iguale a população do país B.")
                print(f"População final de A: {pop_a_temp:.0f}")
                print(f"População final de B: {pop_b_temp:.0f}")

        except ValueError:
            print("Entrada inválida. Por favor, digite apenas números para as populações e taxas.")

        continuar = input("\nDeseja fazer outro cálculo? (s/n): ").lower()
        if continuar != 's':
            break
```

```
def calcular_potencia_manual():
    try:
        base = float(input("Digite o número base: "))
        expoente = int(input("Digite o expoente (número inteiro): "))

        resultado = 1

        if expoente == 0:
            resultado = 1
        elif expoente > 0:
            for _ in range(expoente):
                resultado *= base
        else:
            for _ in range(abs(expoente)):
                resultado *= base
            resultado = 1 / resultado

        print(f"\n{base} elevado a {expoente} é igual a: {resultado}")

    except ValueError:
        print("Entrada inválida. Por favor, digite números válidos para a base e um número inteiro para o expoente.")
    except ZeroDivisionError:
        print("Erro: Não é possível dividir por zero (base 0 com expoente negativo).")

calcular_potencia_manual()
```

```
try:
    numero = int(input("Digite um número inteiro positivo para calcular o fatorial: "))

    if numero < 0:
        print("O fatorial não é definido para números negativos.")
    elif numero == 0:
        print("O fatorial de 0 é 1.")
    else:
        fatorial = 1
        expressao = []
        for i in range(1, numero + 1):
            fatorial *= i
            expressao.append(str(i))

        # Inverte a lista para mostrar a expressão em ordem decrescente (ex: 5.4.3.2.1)
        expressao.reverse()
        print(f"{numero}! = {' . '.join(expressao)} = {fatorial}")
except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
```

```
import math

try:
    numero = int(input("Digite um número inteiro para verificar se é primo: "))

    if numero <= 1:
        print(f"O número {numero} NÃO é primo.")
    elif numero == 2:
        print(f"O número {numero} É primo.")
    elif numero % 2 == 0:
        print(f"O número {numero} NÃO é primo. Ele é divisível por 2.")
    else:
        eh_primo = True
        divisores = []
        for i in range(3, int(math.sqrt(numero)) + 1, 2):
            if numero % i == 0:
                eh_primo = False
                divisores.append(i)
                if i * i != numero:
                    divisores.append(int(numero / i))
                break
        if eh_primo:
            print(f"O número {numero} É primo.")
        else:
            divisores_ordenados = sorted(list(set(divisores)))
            print(f"O número {numero} NÃO é primo. Ele é divisível por 1, {', '.join(map(str, divisores_ordenados))}, e {numero}.")

except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
```

```
import math

def encontrar_primos_com_divisoes():
    total_divisoes = 0
    primos_encontrados = []

    try:
        N = int(input("Digite um número inteiro N para encontrar os primos entre 1 e N: "))

        if N < 2:
            print(f"Não há números primos no intervalo de 1 a {N}.")
            print(f"Total de divisões executadas: {total_divisoes}")
            return

        print(f"Verificando números primos de 1 a {N}...")

        for num in range(2, N + 1):
            is_prime = True

            if num == 2:
                primos_encontrados.append(num)
                continue
            if num % 2 == 0:
                is_prime = False
                # Uma 'divisão' implícita para o check de paridade
                # Poderíamos contar 'num % 2' como uma divisão aqui, mas para foco
                # em divisões dentro do loop, vou omitir esta para não confundir com o sqrt otimizado.
                # Se a intenção é CADA operação de módulo, esta deveria ser adicionada.
                continue

            # Só verifica divisores ímpares a partir de 3 até a raiz quadrada
            limit = int(math.sqrt(num))
            for divisor in range(3, limit + 1, 2):
                total_divisoes += 1 # Conta cada operação de divisão
                if num % divisor == 0:
                    is_prime = False
                    break

            if is_prime:
                primos_encontrados.append(num)

        print(f"\nNúmeros primos encontrados entre 1 e {N}: {primos_encontrados}")
        print(f"Total de divisões executadas para encontrar os primos: {total_divisoes}")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

encontrar_primos_com_divisoes()
```

```
def calcular_media_notas():
    try:
        N = int(input("Quantas notas você deseja inserir para calcular a média? "))

        if N <= 0:
            print("Por favor, insira um número positivo para a quantidade de notas.")
            return

        soma_notas = 0
        print(f"\nDigite as {N} notas:")

        for i in range(1, N + 1):
            nota = float(input(f"Digite a {i}ª nota: "))
            soma_notas += nota

        media = soma_notas / N

        print(f"\nA soma total das notas é: {soma_notas:.2f}")
        print(f"A média aritmética das {N} notas é: {media:.2f}")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro para a quantidade de notas e números para as notas.")

calcular_media_notas()
```

```
def classificar_turma_por_idade_media():
    try:
        n_pessoas = int(input("Quantas pessoas há na turma? "))

        if n_pessoas <= 0:
            print("Por favor, insira um número positivo de pessoas.")
            return

        soma_idades = 0
        print(f"\nDigite a idade de cada uma das {n_pessoas} pessoas:")

        for i in range(1, n_pessoas + 1):
            idade = int(input(f"Digite a idade da {i}ª pessoa: "))
            if idade < 0:
                print("Idade não pode ser negativa. Por favor, digite novamente.")
                i -= 1
                continue
            soma_idades += idade

        media_idade = soma_idades / n_pessoas

        if 0 <= media_idade <= 25:
            classificacao = "Jovem"
        elif 26 <= media_idade <= 60:
            classificacao = "Adulta"
        else:
            classificacao = "Idosa"

        print(f"\nA média de idade da turma é: {media_idade:.2f} anos.")
        print(f"A turma é classificada como: {classificacao}.")

    except ValueError:
        print("Entrada inválida. Por favor, digite apenas números inteiros para o número de pessoas e para as idades.")

classificar_turma_por_idade_media()
```

```
votos_candidato1 = 0
votos_candidato2 = 0
votos_candidato3 = 0

print("\n--- Simulação de Eleição ---\n")

try:
    total_eleitores = int(input("Digite o número total de eleitores: "))

    if total_eleitores <= 0:
        print("O número de eleitores deve ser um valor positivo.")
    else:
        for i in range(1, total_eleitores + 1):
            while True:
                try:
                    voto = int(input(f"Eleitor {i}: Digite seu voto (1, 2 ou 3): "))
                    if voto == 1:
                        votos_candidato1 += 1
                        break
                    elif voto == 2:
                        votos_candidato2 += 1
                        break
                    elif voto == 3:
                        votos_candidato3 += 1
                        break
                    else:
                        print("Voto inválido. Por favor, digite 1, 2 ou 3.")
                except ValueError:
                    print("Entrada inválida. Por favor, digite um número (1, 2 ou 3).")

        print("\n--- Resultado da Eleição ---")
        print(f"Candidato 1: {votos_candidato1} votos")
        print(f"Candidato 2: {votos_candidato2} votos")
        print(f"Candidato 3: {votos_candidato3} votos")

except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro para o total de eleitores.")
```
